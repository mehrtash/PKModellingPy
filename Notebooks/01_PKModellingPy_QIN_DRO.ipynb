## Reading Image Data and Converting to Numpy array

```python
dataDir = "../Data/QIBA_v8_Tofts_4s_0s/"
outputFolder = "./output/QIBA_v8_Tofts_4s_0s/"
# Reading the DCE-MRI image
image = sitk.ReadImage(dataDir + "QIBA_v8_Tofts_4s_0s.nrrd")
# Reading the Label map
label = sitk.ReadImage(dataDir + "label.nrrd")
# Getting the numpy arrays from the images
ndaImage = sitk.GetArrayFromImage(image)
ndaLabel = sitk.GetArrayFromImage(label)
```

## Reading DCE-MRI image meta-data

The DCE-MRI DICOM to NRRD conversion is done in 3D Slicer by reading the DICOM files with Multivolume image plugin [reference] and saving the image volume node as nrrd.<br>
The parameters we need for PK Modellin are:
* Sampling Rate
* Flip Angle (FA)
* Repetition Time (TR)

```python
frameLabels = image.GetMetaData('MultiVolume.FrameLabels')
frameLabelsArray = np.array([x.strip() for x in frameLabels.split(',')],dtype=float)
frameLabelsDiff = np.ediff1d(frameLabelsArray)
# Todo: Check that all differences are the same
samplingRate = frameLabelsDiff[0]
if  image.GetMetaData('MultiVolume.FrameIdentifyingDICOMTagUnits') == 'ms':
    samplingRate = samplingRate/1000
else:
    print 'Warning: Check the frames time units. The units are not in miliseconds'
FA = float(image.GetMetaData('MultiVolume.DICOM.FlipAngle'))
TR = float(image.GetMetaData('MultiVolume.DICOM.RepetitionTime'))
print 'Sampling Rate:', samplingRate, 'FA: ', FA, 'TR: ', TR
```

# Creating AIF

```python
m,n,p,q = np.shape(ndaImage)
# Make the AIF signal
# For now it uses the first one change it to a more sophisticated one
meanSp = np.zeros(q)
fig = figure(num=None, figsize=(16, 3), dpi=80, facecolor='w', edgecolor='k')
for i in xrange(0,m):
    for j in xrange(0,n):
        for k in xrange(0,p):
            # The AIF is marked as 3 in the label map
            if ndaLabel[i,j,k] == 3:
                sp = ndaImage[i,j,k,:]
                ax1 = plt.subplot(1,2,1)
                plot(sp)
                meanSp = (meanSp + ndaImage[i,j,k,:])/2
ax1.grid()
ax1.set_title('AIF Signals')
ax1.set_xlabel('Time Points')
ax1.set_ylabel('MR Signal')
converter = SignalToConcentrationConverter()
converter.signal = meanSp
converter.BAT = 10
converter.T10 = 1440
converter.Relaxivity = 0.0045
converter.FA = FA
converter.TR = TR
aif = converter.convert()
ax2 = plt.subplot(1,2,2)
ax2.grid()
ax2.set_title('AIF Concentration')
ax2.set_xlabel('Time Points')
ax2.set_ylabel('Concentration mmol/l')
plot(aif,'b-o')
```

## PK Modelling

```python
bc = BATCalculator()
# Solve PK modelling
solver = PKSolver()
ndaKtrans = np.zeros((m,n,p))
ndaVe = np.zeros((m,n,p))
ndaRsquared = np.zeros((m,n,p))
ndaBATCt = np.zeros((m,n,p))
ndaCtConcentration = np.zeros((m,n,p,q))
for i in xrange(0,m):
    for j in xrange(0,n):
        for k in xrange(0,p):
            if ndaLabel[i,j,k] == 1:
                st = ndaImage[i,j,k,:] # Tissue Signal
                # Tissue signal to concentration conversion
                converter.signal = st
                converter.FA = FA
                converter.TR = TR
                converter.T10 = 1000
                converter.BAT = 10
                converter.Relaxivity = 0.0045
                TRF = converter.convert()
                # Assigning the concentration curve
                ndaCtConcentration[i,j,k,:] = TRF
                solver.AIF = aif
                solver.TRF = TRF
                solver.BATCp = 15 
                solver.BATCt = 16
                ndaBATCt[i,k,k] = solver.BATCt
                solver.samplingRate = samplingRate
                solver.hematocrit = 0.45
                result = solver.solve()
                ndaKtrans[i,j,k] = result[0]
                ndaVe[i,j,k] = result[1]
                ndaRsquared[i,j,k] = solver.calculateGoodnessOfFit()
```

## Saving Ouput to File

```python
# Saving outputs
# ktrans
ktransImage = sitk.GetImageFromArray(ndaKtrans)
ktransImage.CopyInformation(label)
sitk.WriteImage(ktransImage,outputFolder+'ktrans.nrrd')
# ve
veImage = sitk.GetImageFromArray(ndaVe)
veImage.CopyInformation(label)
sitk.WriteImage(veImage,outputFolder+'ve.nrrd')
# r_squared
rSquaredImage = sitk.GetImageFromArray(ndaRsquared)
rSquaredImage.CopyInformation(label)
sitk.WriteImage(rSquaredImage,outputFolder+'rSquared.nrrd')
# BAT CT
BATCTImage = sitk.GetImageFromArray(ndaBATCt)
BATCTImage.CopyInformation(label)
sitk.WriteImage(BATCTImage,outputFolder+'BATCt.nrrd')
## concentration
imageCtConcentration = sitk.GetImageFromArray(ndaCtConcentration)
imageCtConcentration.CopyInformation(label)
sitk.WriteImage(image,outputFolder+'concentration.nrrd')
```
